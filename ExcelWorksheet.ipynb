```python
import pandas as pd
```

1. Employee_Master.xlsx

```python
df = pd.read_excel('Excel Files/Employee_Master (1).xlsx', sheet_name='Sheet1')
df.head(10)
```

```python
df.drop_duplicates(inplace=True)
df.fillna({"Age":df["Age"].mean()}, inplace=True)
df.fillna({"Salary":df["Salary"].mean()}, inplace=True)
df["Salary"].max()
df["Salary"].min()
df.groupby("Department")["Salary"].sum()
```

```python
new_df = df[df["Department"]=="IT"]
new_df.to_excel("ITdepartment.xlsx", sheet_name="IT", index=False)
```

2. Sales_Transactions.xlsx

```python
df = pd.read_excel('Excel Files/Sales_Transactions.xlsx', sheet_name='Sheet1')
df["Product"].unique()
df.head()
```

```python
df["Total_Value"].sum()
df.groupby("Product").agg(Total_Revenue = ("Total_Value","sum")).reset_index()
df.groupby("Product")["Total_Value"].sum().max()
df[df["Qty"]>3]
```

3. Attendance Report

```python
df = pd.read_excel("Excel Files/Attendance_Report.xlsx", sheet_name="Sheet1")
df.head()
```

```python
df["Total_WorkingDays"] = df["Days_Present"] + df["Days_Absent"]
df["Attendance%"] = ((df["Days_Present"] / df["Total_WorkingDays"])*100).round(2)
df[df["Attendance%"] < 75]
df.sort_values(by="Attendance%", ascending=False)
```

4. Product Inventory 

```python
df = pd.read_excel("Excel Files/Product_Inventory.xlsx")
df.head()
```

```python
df["Stock"].sum()
df["Unit_Price"].max()
df["Unit_Price"].min()
df[df["Stock"] < 50]
df.groupby("Category").agg(Total_Stock = ("Stock","sum")).reset_index()
```

5. Student Marks


```python
df = pd.read_excel("Excel Files/Student_Marks.xlsx")
df.head()
```

```python
df["Avarage_Marks"] = df[["Maths","Science","English"]].mean(axis=1).round(1)
df["Total_Marks"] = df[["Maths","Science","English"]].sum(axis=1)
df.sort_values(by="Total_Marks",ascending=False).head(3)
df[["Maths","Science","English"]].mean().idxmin()
```
